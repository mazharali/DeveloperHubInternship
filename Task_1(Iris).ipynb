```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# data = sns.load_dataset('iris')
data = pd.read_csv('iris.csv')
data.head(10)
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='petal_length', y ='petal_width', hue='species', data=data)
plt.title('Scatter Plot of Patel Length vs Petal Width')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width cm')
# plt.grid(True)
plt.show()
```

```python
plt.hist(data['sepal_length'], bins=15)
plt.title('Histogram of Sepal Length Distribution')
plt.xlabel('Sepel Length cm')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.boxplot(x='species', y='sepal_width', data=data)
plt.title('Box plot of species by its width')
plt.xlabel('Species')
plt.ylabel('Sepal width')
plt.show()
```
